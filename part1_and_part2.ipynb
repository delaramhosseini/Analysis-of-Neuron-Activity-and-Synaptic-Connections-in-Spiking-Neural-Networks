```python
import pymonntorch as pmt

import torch
import matplotlib.pyplot as plt
```

```python
from timeresolution import Timeresolution
from current import ConstanceCurrent,UniformCurrent
from lif import LIF
from inputforward import FullConnectivityFirstOption, FullConnectivitySecondOption, Scaling, FixedAAndFinite
from dandrite import Dandrite
from populationactivity import PopulationActivity
from plot import print_plot
```

```python
def BuildNeuronNetwork_SingleNeuron(model=LIF(tau=10,
                                                R=1.5,
                                                u_rest=-67,
                                                u_reset=-75,
                                                threshold=-55,
                                                ratio=1), 
                                    current=ConstanceCurrent(current = 8.5), 
                                    dt = 1, 
                                    iteration_num = 100, 
                                    PLT = True, 
                                    size=5,
                                    InputForward = FullConnectivityFirstOption(J0=20)):
                    
    net1 = pmt.Network(behavior = {1: Timeresolution(dt = dt)} )
    
    pop1 = pmt.NeuronGroup(size = size,
                    net = net1,
                    behavior = {
                        3 : current,
                        5 : Dandrite(),
                        6 : model,
                        7 : PopulationActivity(),
                        13 : pmt.Recorder(variables=["u","I","w","T"],tag="ng1_rec"),
                        14 : pmt.EventRecorder( variables=["spike"],tag="ng1_even")
                        })
    
    syn_inside_ng = pmt.SynapseGroup(tag="synaps_inside_neurongroup",
                                    net= net1,
                                    src= pop1,
                                    dst= pop1,
                                    behavior={4: InputForward})
    
    net1.initialize()
    net1.simulate_iterations(iteration_num)

    if PLT: print_plot(net1,ngs=[pop1])
    
    return net1
```

```python
def BuildNeuronNetwork_2NeoutonGroup(modelـpop1=LIF(tau=10,
                                                R=1.5,
                                                u_rest=-67,
                                                u_reset=-75,
                                                threshold=-55,
                                                ratio=1), 
                                    modelـpop2=LIF(tau=10,
                                                R=1.5,
                                                u_rest=-67,
                                                u_reset=-75,
                                                threshold=-55,
                                                ratio=1),
                                    current=ConstanceCurrent(current = 8.5), 
                                    dt = 0.25, 
                                    iteration_num = 100, 
                                    PLT = True, 
                                    size=10,
                                    InputForward_pop1_to_pop2 = FullConnectivityFirstOption(J0=10),
                                    InputForward_pop2_to_pop1 = FullConnectivityFirstOption(J0=10),
                                    InputForward_pop1_to_pop1 = FullConnectivityFirstOption(J0=5),
                                    InputForward_pop2_to_pop2 = FullConnectivityFirstOption(J0=5)):
                    
    net1 = pmt.Network(behavior = {1: Timeresolution(dt = dt)} )
    
    pop1 = pmt.NeuronGroup(size = size,
                            net = net1,
                            behavior = {
                                3 : current,
                                5 : Dandrite(),
                                6 : modelـpop1,
                                7 : PopulationActivity(),
                                13 : pmt.Recorder(variables=["u","I","w","T"],tag="ng1_rec"),
                                14 : pmt.EventRecorder( variables=["spike"],tag="ng1_even")
                                })
            
    pop2 = pmt.NeuronGroup(size = size,
                    net = net1,
                    behavior = {
                        3 : ConstanceCurrent(current = 0),
                        5 : Dandrite(),
                        6 : modelـpop2,
                        7 : PopulationActivity(),
                        13 : pmt.Recorder(variables=["u","I","w","T"],tag="ng1_rec"),
                        14 : pmt.EventRecorder( variables=["spike"],tag="ng1_even")
                        })
    
    syn_from_pop1_to_pop2 = pmt.SynapseGroup(tag="synaps_from_pop1_to_pop2",
                                    net= net1,
                                    src= pop1,
                                    dst= pop2,
                                    behavior={4: InputForward_pop1_to_pop2})
    
    # syn_from_pop2_to_pop1 = pmt.SynapseGroup(tag="synaps_from_pop2_to_pop1",
    #                                 net= net1,
    #                                 src= pop2,
    #                                 dst= pop1,
    #                                 behavior={4: InputForward_pop2_to_pop1})
    
    # syn_inside_pop1 = pmt.SynapseGroup(tag="synaps_inside_pop1",
    #                                 net= net1,
    #                                 src= pop1,
    #                                 dst= pop1,
    #                                 behavior={4: InputForward_pop1_to_pop1})
    
    # syn_inside_pop2 = pmt.SynapseGroup(tag="synaps_inside_pop2",
    #                                 net= net1,
    #                                 src= pop2,
    #                                 dst= pop2,
    #                                 behavior={4: InputForward_pop2_to_pop2})
    
    net1.initialize()
    net1.simulate_iterations(iteration_num)

    if PLT: print_plot(net1,ngs=[pop1, pop2])
    # if PLT: Print_Plot(net1)
    
    return net1
```
